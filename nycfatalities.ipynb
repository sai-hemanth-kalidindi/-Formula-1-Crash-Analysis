```python
import pandas as pd
from scipy.stats import ttest_ind

# Convert crash time to hour
df['CRASH_HOUR'] = pd.to_datetime(df['CRASH TIME'], format='%H:%M', errors='coerce').dt.hour

# Create time groups
df['TIME_PERIOD'] = df['CRASH_HOUR'].apply(lambda x: 'Night' if 0 <= x < 6 else 'Day')

# Create binary flag for fatal crashes
df['FATAL_FLAG'] = df['NUMBER OF PERSONS KILLED'].apply(lambda x: 1 if x > 0 else 0)

# Separate night and day groups
night_fatal = df[df['TIME_PERIOD'] == 'Night']['FATAL_FLAG']
day_fatal = df[df['TIME_PERIOD'] == 'Day']['FATAL_FLAG']

# T-test
t_stat, p_value = ttest_ind(night_fatal, day_fatal, equal_var=False)

# Output
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("✅ Reject the null hypothesis: Fatal crashes are more likely at night.")
else:
    print("❌ Fail to reject the null hypothesis: No significant difference found.")

```

```python
# Create a new binary column for alcohol involvement
df['ALCOHOL_INVOLVED'] = df['CONTRIBUTING FACTOR VEHICLE 1'].str.lower().str.contains('alcohol').fillna(False)

# Group crashes by alcohol involvement
alcohol_fatal = df[df['ALCOHOL_INVOLVED'] == True]['FATAL_FLAG']
non_alcohol_fatal = df[df['ALCOHOL_INVOLVED'] == False]['FATAL_FLAG']

# Perform t-test
t_stat_alc, p_val_alc = ttest_ind(alcohol_fatal, non_alcohol_fatal, equal_var=False)

# Output
print(f"T-statistic: {t_stat_alc}")
print(f"P-value: {p_val_alc}")

# Interpretation
if p_val_alc < 0.05:
    print("✅ Reject the null hypothesis: Fatal crashes are more likely when alcohol is a contributing factor.")
else:
    print("❌ Fail to reject the null hypothesis: No significant difference found for alcohol involvement.")
```

```python
from scipy.stats import f_oneway

# Drop missing boroughs
df_boroughs = df.dropna(subset=['BOROUGH'])

# Group fatality data by borough
borough_groups = [group['NUMBER OF PERSONS KILLED'].dropna()
                  for name, group in df_boroughs.groupby('BOROUGH')]

# Run ANOVA test
f_stat, p_val = f_oneway(*borough_groups)

# Output
print(f"F-statistic: {f_stat}")
print(f"P-value: {p_val}")

# Interpretation
if p_val < 0.05:
    print("✅ Reject the null hypothesis: Fatalities per crash vary significantly across boroughs.")
else:
    print("❌ Fail to reject the null hypothesis: No significant difference found between boroughs.")
```

```python
from scipy.stats import ttest_ind

# Extract hour from 'CRASH TIME' and create IS_NIGHT column
df['CRASH_HOUR'] = pd.to_datetime(df['CRASH TIME'], errors='coerce').dt.hour
df['IS_NIGHT'] = df['CRASH_HOUR'].between(0, 6).astype(int)

# Separate groups
night_fatalities = df[df['IS_NIGHT'] == 1]['NUMBER OF PERSONS KILLED']
day_fatalities = df[df['IS_NIGHT'] == 0]['NUMBER OF PERSONS KILLED']

# Run independent T-test
t_stat, p_value = ttest_ind(night_fatalities, day_fatalities, equal_var=False)

print("T-statistic:", t_stat)
print("P-value:", p_value)

# Interpretation
if p_value < 0.05:
    print("✅ Reject the null hypothesis: Fatal crashes are more likely at night.")
else:
    print("❌ Fail to reject the null hypothesis.")
```

```python
from scipy.stats import ttest_ind

# Extract hour safely
df['CRASH_HOUR'] = pd.to_datetime(df['CRASH TIME'], errors='coerce').dt.hour

# Filter valid rows
df_valid = df.dropna(subset=['CRASH_HOUR', 'NUMBER OF PERSONS KILLED'])

# Create IS_NIGHT flag
df_valid['IS_NIGHT'] = df_valid['CRASH_HOUR'].between(0, 6).astype(int)

# Create groups
night_fatalities = df_valid[df_valid['IS_NIGHT'] == 1]['NUMBER OF PERSONS KILLED']
day_fatalities = df_valid[df_valid['IS_NIGHT'] == 0]['NUMBER OF PERSONS KILLED']

# T-test
t_stat, p_value = ttest_ind(night_fatalities, day_fatalities, equal_var=False)

print("T-statistic:", t_stat)
print("P-value:", p_value)

if p_value < 0.05:
    print("✅ Reject the null hypothesis: Fatal crashes are more likely at night.")
else:
    print("❌ Fail to reject the null hypothesis.")
```
